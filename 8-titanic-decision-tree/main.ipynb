```python
import pandas as pd
```

```python
df = pd.read_csv('dataset/titanic.csv')
df.head()
```

```python
inputs = df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'], axis='columns')
inputs
```

```python
inputs_clean = inputs.dropna()
```

```python
inputs_clean
```

```python
targets = inputs_clean['Survived']
targets
```

```python
inputs_n = inputs_clean.drop(['Survived'], axis='columns')
```

```python
inputs_n
```

```python
targets
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_sex = LabelEncoder()
```

```python
inputs_n['sex_n'] = le_sex.fit_transform(inputs_n['Sex'])

inputs_n
```

```python
inputs_fix = inputs_clean.drop(['Sex'],axis='columns')
inputs_fix.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(inputs_fix,targets, test_size=0.2,random_state=2022)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```

```python
model.score(X_test,y_test)
```

```python
import joblib
```

```python
joblib.dump(model,'model')
```
