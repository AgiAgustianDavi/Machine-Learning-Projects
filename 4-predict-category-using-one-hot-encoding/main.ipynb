```python
import pandas as pd
```

```python
df = pd.read_csv('dataset/homeprices.csv')
df.head()
```

```python
dumies = pd.get_dummies(df.town)
dumies
```

```python
df_merged = pd.concat([df,dumies], axis='columns')
df_merged
```

ketika kita membuat dummy, kita harus mengurangi 1 kolom supaya ML aman. contoh jika kita menghasilkan 4 kolom dummy, maka kita harus hapus 1 kolom, bebas yang mana, tapi mending hapus yang paling ujung

```python
df_final = df_merged.drop(['town','west windsor'], axis='columns')
df_final
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
X = df_final.drop('price', axis='columns')
X
```

```python
y = df_final.price
y
```

```python
model.fit(X,y)
```

```python
model.predict([[2800,0,1]])
```

```python
model.predict([[3400,0,0]])
```

```python
model.score(X,y)
```

### One-hot Encoding

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
```

```python
df_le = df
```

```python
df_le.town = le.fit_transform(df_le.town)
```

```python
df
```

```python
X = df_le[['town','area']].values
X
```

```python
y = df_le.price
y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('town', OneHotEncoder(),[0])], remainder='passthrough')
```

```python
X = ct.fit_transform(X)
```

```python
X
```

```python
X = X[:,1:]
X
```

```python
model.fit(X,y)
```

```python
model.predict([[1,0,2800]]) # 2800 sqr ft home in robbinsville
```

```python
model.predict([[0,1,3400]]) # 3400 sqr ft home in west windsor
```
